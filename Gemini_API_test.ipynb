```python
import json
import requests
import base64
from io import BytesIO
from PIL import Image
import matplotlib.pyplot as plt

with open(r"C:\Users\user\Downloads\animal.jpg", "rb") as image_file:
    image_base64_string = base64.b64encode(image_file.read()).decode('utf-8')
    
image_data = base64.b64decode(image_base64_string)
image = Image.open(BytesIO(image_data))

plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
url = f'https://generativelanguage.googleapis.com/v1/models/gemini-1.5-flash:generateContent?key={api_key}'
headers = {'Content-Type': 'application/json'}
data = {
    "contents": [
        {
            "parts": [
                {"text": "詳細說明你在這張圖片中看到什麼？"},
                {
                    "inline_data": {
                        "mime_type": "image/jpeg",
                        "data": image_base64_string
                    }
                }
            ]
        },
    ]
}
response = requests.post(url, headers=headers, json=data)
print(f"response status_code: {response.status_code}")
response_json = response.json()
description = response_json['candidates'][0]['content']['parts'][0]['text']
print(description)
```
